```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
from datetime import timedelta
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
cities_df = pd.read_csv('Resources/cities.csv')
cities_df.head()
```

```python
# Plot a scattergram showing Lat vs Humidity
x_values = cities_df['Lat']
y_values = cities_df['Humidity']
plt.scatter(x_values,y_values, edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Cities")
plt.xlabel('Latitude')
plt.ylabel('Humidity')
plt.title('City Latitude vs Humidity ("01/05/17")')
plt.grid(color="gray",linestyle = "solid")
plt.savefig("Images/Humidity.png")
plt.show()
```
